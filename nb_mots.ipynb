## Analyse des bases des labels séparemment

```python
import pandas as pd
```

```python
data = pd.read_csv('train.csv')
data
```

### Positifs

```python
data_posi = data.loc[data['sentiment'] == "positive", :]
selected_text_posi = data_posi['selected_text']
selected_text_posi = selected_text_posi.to_numpy()
vecteur_posi, d = vectorize(selected_text_posi)
```

```python
len(d)
```

### Negatifs

```python
data_neg = data.loc[data['sentiment'] == "negative", :]
selected_text_neg = data_neg['selected_text']
selected_text_neg = selected_text_neg.to_numpy()
vecteur_neg, d0 = vectorize(selected_text_neg)
len(d0)
```

### Neutres

```python
data_neu = data.loc[data['sentiment'] == "neutral", :]
selected_text_neu = data_neu['selected_text']
selected_text_neu = selected_text_neu.to_numpy()
vecteur_neu, d1 = vectorize(selected_text_neu)
len(d1)
```

### Union positifs et négatifs

```python
data_sen = data.loc[data['sentiment'] != "neutral", :]
selected_text_sen = data_sen['selected_text']
selected_text_sen = selected_text_sen.to_numpy()
vecteur_sen, d2 = vectorize(selected_text_sen)
len(d2)
```
